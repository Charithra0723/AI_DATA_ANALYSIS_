```python
# Seaborn Histogram with KDE
# Description: Create a histogram using Seaborn that i
# data: [23, 45, 56, 78, 89, 90, 67, 88, 76, 59].
import seaborn as sns
import matplotlib.pyplot as plt
marks=[23,45,56,78,89,90,67,88,76,59]
sns.histplot(marks,kde=True,bins=5,color='skyblue',edgecolor='black')
plt.title('Seaborn histogram with kde')
plt.xlabel('Marks')
plt.ylabel('Freq')
plt.show()
plt.close()
```
