```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.
import matplotlib.pyplot as plt
product=['A','B','C','D','E']
value=[23,75,57,90,65]
plt.bar(product,value,color='pink')
plt.title('Number of sales',fontsize=12,color='darkred')
plt.xlabel('Products',fontsize=10,color='lightblue')
plt.ylabel('Values',fontsize=10,color='lightblue')
plt.grid(True)
plt.show()
plt.close()
```
