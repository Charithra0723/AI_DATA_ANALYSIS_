```python
# Scatter Plot of Variable Relationships
# Description: Use Seaborn to create a scatter plot showing the relationship between two
# variables.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df=pd.DataFrame({
    'Subject':['A','B','C','D','E'],
    'Marks':[43,76,83,50,84]
})
if df[['Subject','Marks']].isnull().values.any():
    raise ValueError("Missing values found in data")
else:
    sns.scatterplot(data=df,x='Subject',y='Marks')
    plt.title('Marks of a subject')
    plt.xlabel('Subject')
    plt.ylabel('Marks')
    plt.show()
```
