```python
# Matplotlib Subplots for Multiple Data Sets
# Description: Create subplots using Matplotlib to display the bar plot for product sales and the scatter plot for
# height vs. weight side by side.
import matplotlib.pyplot as plt
product=['A','B','C','D','E']
value=[23,75,57,90,65]
heights=[160,170,165,180,175]
weights=[55,65,60,80,76]
fig, axs=plt.subplots(1,2,figsize=(12,5))
axs[0].bar(product,value,color='blue')
axs[0].set_title('Bar plot')
axs[0].set_xlabel('product')
axs[0].set_ylabel('Value')

axs[1].scatter(heights,weights,color='teal')
axs[0].set_title('Scatter plot')
axs[0].set_xlabel('Height(in cm)')
axs[0].set_ylabel('Weight')
```
