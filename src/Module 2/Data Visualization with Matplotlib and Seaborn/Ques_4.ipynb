```python
# Basic Seaborn Bar Plot
# Description: Using Seaborn, plot the same bar plot from Question 1 to compare the sales data.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df=pd.DataFrame ({'product':['A','B','C','D','E'],
 'value':[23,75,57,90,65]})

sns.barplot(data=df,x='product',y='value')
plt.title('Number of sales')
plt.xlabel('Products')
plt.ylabel('Values')
plt.show()
plt.close()
```
