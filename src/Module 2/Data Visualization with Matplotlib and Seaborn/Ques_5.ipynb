```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weight.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def validate_dataframe(df, required_columns):
    """Validates input DataFrame structure and content."""
    if not isinstance(df, pd.DataFrame):
        raise TypeError("Input must be a pandas DataFrame.")
    if df.empty:
        raise ValueError("DataFrame is empty.")
    for col in required_columns:
        if col not in df.columns:
            raise ValueError(f"Missing required column: {col}")
        if not pd.api.types.is_numeric_dtype(df[col]):
            raise TypeError(f"Column '{col}' must be numeric.")

def plot_height_vs_weight(df):
    """Plots a Seaborn scatter plot for height vs. weight."""
    validate_dataframe(df, ['Height', 'Weight'])

    sns.set(style="whitegrid")
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x='Height', y='Weight', color='blue', s=80)

    plt.title('Height vs. Weight Scatter Plot', fontsize=14)
    plt.xlabel('Height (cm)')
    plt.ylabel('Weight (kg)')
    plt.tight_layout()
    plt.show()

# Sample (can be replaced by large dataset)
data = {
    'Height': [150, 160, 165, 170, 175, 180, 185, 190],
    'Weight': [50, 55, 60, 65, 70, 75, 80, 85]
}
df = pd.DataFrame(data)

# Call plotting function
plot_height_vs_weight(df)
```
