```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weight.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df=pd.DataFrame({'heights':[160,170,165,180,175], 'weights':[55,65,60,80,76]})
sns.set(style="whitegrid")
sns.scatterplot(data=df,x='heights',y='weights',color='blue',s=80)
plt.title('Height vs Weight scatter plot')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```
