```python
# Title: Advanced Data Visualization
# Description: Create a pair plot for the dataset to display pairwise relationships.
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
def load_data():
    try:
        df=sns.load_dataset("iris")
        if df.isnull().values.any():
            print("Warning:Dataset contians missing values.Consider handling before plotting")
        if df.empty:
            raise ValueError("The dataset is empty")
        return df
    except Exception as e:
        print(f"Error loading dataset: {e}")
        return None
def create_pair_plot(df):
    try:
        sns.pairplot(df,hue="species",corner=True)
        plt.suptitle("Pairwise relationship")
        plt.show()
    except Exception as e:
        print(f"Error creating pair plot: {e}")

df=load_data()
if df is not None:
    create_pair_plot(df)
```
