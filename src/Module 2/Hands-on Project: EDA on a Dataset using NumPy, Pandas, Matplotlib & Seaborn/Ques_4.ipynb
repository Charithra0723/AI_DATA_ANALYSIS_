```python
# Title: Basic Data Visualization
# Description: Plot the distribution of one feature using Seaborn.
# Basic Data Visualization
# Description: Plot the distribution of one feature using Seaborn with improvements for clarity, error handling, and modularity.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def plot_feature_distribution(data, feature, bins=20, kde=True):
    
    if feature not in data.columns:
        print(f"Feature '{feature}' not found in dataset.")
        return

    plt.figure(figsize=(8, 5))
    sns.histplot(data[feature], kde=kde, bins=bins)
    plt.title(f'Distribution of {feature.capitalize()}')
    plt.xlabel(feature.capitalize())
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Load dataset with error handling
try:
    df = sns.load_dataset('iris')
    if df.empty:
        raise ValueError("Loaded dataset is empty.")
    plot_feature_distribution(df, 'sepal_length')  # You can change the feature here
except Exception as e:
    print(f"Error loading or processing dataset: {e}")
```
